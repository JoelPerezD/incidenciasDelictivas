# Delitos en mexico hasta abril del 2024

En el siguiente proyecto se realiza la conexion a una base de datos con los datos de https://www.gob.mx/sesnsp/acciones-y-programas/datos-abiertos-de-incidencia-delictiva de la seccion Reportes de incidencia delictiva al mes de abril 2024 (Nueva metodología).

```python
import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt
from tabulate import tabulate
import matplotlib.pyplot as plt  # importamos la liberia para crear graficas 
```

Se realiza la conexion a la base de datos local

```python
def connect_to_db():
    return mysql.connector.connect(
        host="localhost",
        user="root",
        password="12345*",
        database="crimenes"
    )
```

Metodo para ejecutar querys, consultas a la base de datos

```python
def fetch_data(query): # funcion ejecuta querys
    connection = connect_to_db()
    cursor = connection.cursor()
    cursor.execute(query)
    data = cursor.fetchall()
    columns = [col[0] for col in cursor.description]
    df = pd.DataFrame(data, columns=columns)
    cursor.close()
    connection.close()
    return df
```

```python
query = "SELECT * FROM crimenes.indicecriminal where Entidad='sonora'"
data_df = fetch_data(query)
data_df
# comprobamos que se realize correctamente la consulta, en donde podemos revisar los delitos rgistrados en sonora del año 2015 hasta el 2023
```

```python
data_df.info() # sacar la informacion de la tabla para saber que contiene y las consultas que realizaremos
```

en la siguiente tabla sumaremos todos los meses, le colocaremos el titulo a la suma "total delitos" y los agruparemos por entidad , al final los ordenamos por entidad para determinar cuales son los estados con mas delitos del 2015 al 2024. determinamos que el estado de mexico es el estado con mas delitos

```python
query2 ="SELECT Entidad, SUM(Enero + Febrero + Marzo + Abril + Mayo + Junio + Julio + Agosto + Septiembre + Octubre + Noviembre + Diciembre)  AS Total_Delitos FROM crimenes.indicecriminal GROUP BY  Entidad order by Total_Delitos desc"
delitos_por_entidad = fetch_data(query2)
delitos_por_entidad  
```

```python
## crearemos un grafico de los delitos por mes 
delitos_por_mes = data_df[['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']].sum()
# creamos un dataframe con los delitos en una lista 
plt.figure(figsize=(20, 15))
delitos_por_mes.plot(kind='bar', color='skyblue')
plt.title('Delitos por Mes')
plt.xlabel('Mes')
plt.ylabel('Número de Delitos')
plt.xticks(rotation=45)
plt.show()
```

```python
# agrupando por entidad y numero de casos por tipo de delito
agrupacion_entidad= "SELECT  Entidad, tipo_de_delito, COUNT(*) AS Numero_de_Casos FROM crimenes.indicecriminal GROUP BY Entidad, tipo_de_delito"
delitos_por_entidad1 = fetch_data(agrupacion_entidad)
delitos_por_entidad1
```

```python
crimen_mas_comun = "SELECT Año, tipo_de_delito, COUNT(*) AS Numero_de_Casos FROM crimenes.indicecriminal GROUP BY Año, tipo_de_delito ORDER BY Año, Numero_de_Casos DESC;"
crimenes_por_anio= fetch_data(crimen_mas_comun)
crimenes_por_anio
# observamos que el mas comun es el robo durante el 2015, en el 2023 los mas comunes sin falsificacion
```

```python

crimenes_por_anio_sonora ="SELECT Año, SUM(Enero + Febrero + Marzo + Abril + Mayo + Junio + Julio + Agosto + Septiembre + Octubre + Noviembre + Diciembre)  AS Total_Delitos FROM crimenes.indicecriminal WHERE Entidad = 'Sonora' GROUP BY Año ORDER BY Año"

crimenes_sonora= fetch_data(crimenes_por_anio_sonora)
crimenes_sonora
# Sumamos los delitos sumando los campos de todos los meses donde entidad sea igual a sonora
```

```python
# creamos una grafica para validar su ha dismunuido o no el crimen
# Graficar
crimenes_sonora.plot(, stacked=True, figsize=(12, 8))
plt.title('Cantidad de Delitos en Sonora por Año')
plt.xlabel('Año')
plt.ylabel('Número de Casos')
plt.xticks(rotation=45)
plt.legend(title='Tipo de Delito')
plt.show()
```
