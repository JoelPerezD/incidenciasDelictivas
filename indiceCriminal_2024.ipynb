Metodo para ejecutar querys, consultas a la base de datos

```python
def fetch_data(query): # funcion ejecuta querys
    connection = connect_to_db()
    cursor = connection.cursor()
    cursor.execute(query)
    data = cursor.fetchall()
    columns = [col[0] for col in cursor.description]
    df = pd.DataFrame(data, columns=columns)
    cursor.close()
    connection.close()
    return df
```

```python
query = "SELECT * FROM crimenes.indicecriminal where Entidad='sonora'"
data_df = fetch_data(query)
data_df
# comprobamos que se realize correctamente la consulta, en donde podemos revisar los delitos rgistrados en sonora del año 2015 hasta el 2023
```
